```python
#import library
import pandas as pd

# Membaca file Excel
df = pd.read_excel("data-sell.xlsx", engine='openpyxl')

# Menampilkan 5 baris pertama
df.head()
```

Eksplorasi Awal dengan EDA Sederhana

```python
# Ukuran data
print("Jumlah baris dan kolom:", df.shape)

# Nama kolom
print("Kolom:\n", df.columns)

# Info tipe data
df.info()

# Statistik deskriptif
df.describe(include='all')
```

Data Cleaning

```python
# Cek missing value
df.isnull().sum()
```

```python
#Baris dengan nilai NaN
df.dropna(inplace=True)
```

```python
# Menampilkan seluruh baris duplikat berdasarkan semua kolom
duplicates_all = df[df.duplicated(keep=False)]
print("Data duplikat berdasarkan semua kolom:")
display(duplicates_all)
```

```python
total_duplicates = df.duplicated().sum()
print(f"Jumlah baris duplikat: {total_duplicates}")
```

Penyesuaian type data

```python
#Mengubah kolom ke tipe tanggal / datetime
df['tgl_trs'] = pd.to_datetime(df['tgl_trs'], errors='coerce')
```

```python
#Mengubah kolom numerik ke tipe int
df['SKU'] = df['SKU'].astype(int)
df['qty'] = df['qty'].astype(int)
df['net_sales'] = df['net_sales'].astype(int)
df['gm'] = df['gm'].astype(int)
```

```python
#pada kolom %gm menghapus tanda "%" dan mengubah ke float
df['%gm'] = df['%gm'].astype(str).str.replace('%', '').astype(float)
```

```python
# Membaca file Excel
df = pd.read_excel("data-sell.xlsx", engine='openpyxl')

# Menampilkan 5 baris pertama
df.head()
```

Selesai, selanjutnya download dalam bentuk CSV dengan menghapus header

```python
# Simpan ke CSV tanpa header dan tanpa index
df.to_csv('data_cleaned.csv', index=False, header=False)

# Unduh file menggunakan Google Colab helper
from google.colab import files
files.download('data_cleaned.csv')
```

```python
import gzip
import shutil

# Nama file CSV dan GZIP
csv_filename = 'data_cleaned.csv'
gzip_filename = 'data_penjualan.csv.gz'

# Kompres file
with open(csv_filename, 'rb') as f_in:
    with gzip.open(gzip_filename, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```
